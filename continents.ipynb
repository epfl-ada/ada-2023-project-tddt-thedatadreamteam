### Getting the categories by Geography

```python
def assign_continent(categories,continent,df_categories):
    filter = df_categories[df_categories['Categories'].isin(categories)]
    articles = list(filter['Articles'].values)
    for article in articles:
        continents[article] = continent
    df_categories = df_categories[~df_categories['Articles'].isin(articles)]
    return df_categories
```

```python
#If the title of the article has as a word the name of e.g France directly to Europe
new_articles = []
for article in articles_left:
    for word in continents.keys():
        if word in article:
            continents_all[article] = continents[word]
            new_articles.append(article)
    if 'Canadian' in article or 'American' in article:
        continents_all[article] = 'North_America'
        new_articles.append(article)
    if 'British' in article or 'European' in article:
        continents_all[article] = 'Europe'
        new_articles.append(article)
```
